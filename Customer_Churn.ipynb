```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/Customer-Churn-Prediction-ML/Churn_Modelling.csv')
```

```python
data.head()
```

```python
data.info()
```

**Choosing the columns that need to do this prediction**

```python
X= data.iloc[:,3:13]
y= data.iloc[:,13]
```

```python
X.head ()
```

```python
y.head()
```

```python
print(X['Geography'].unique())
print(X['Gender'].unique())
print(X['NumOfProducts'].unique())
print(X['HasCrCard'].unique())
print(X['IsActiveMember'].unique())
```

```python
print(X['Geography'].value_counts())
print(X['Gender'].value_counts())
print(X['NumOfProducts'].value_counts())
print(X['HasCrCard'].value_counts())
print(X['IsActiveMember'].value_counts())
```

```python
X.describe()
```

**To confirm there is no null value**

```python
X.isnull().sum()
```

**Check datatypes again**

```python
X.dtypes
```

```python
X= pd.get_dummies(X)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

```

```python
# Assuming X is your data and y are your labels
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, stratify= y,random_state=0)
```

```python
print(X_train)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 100)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
cm= confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
import pickle
pickle.dump(rf, open('model.pkl', 'wb'))
```
